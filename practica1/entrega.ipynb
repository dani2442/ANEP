# 2. Problema mal puestos

## a)


Tenemos:
\begin{equation}
\frac{\partial u}{\partial t}(x,y,t) = a e^{at}\sin(\pi x)\sin(\pi y)\quad\quad
\frac{\partial^2 u}{\partial x^2}(x,y,t) = -\pi^2 e^{at}\sin(\pi y)\sin(\pi x) \quad\quad
\frac{\partial^2 u}{\partial y^2}(x,y,t) = -\pi^2 e^{at}\sin(\pi y)\sin(\pi x) \quad\quad
\end{equation}

Usando la primera condicin:
\begin{equation}
a e^{at}\sin(\pi x)\sin(\pi y) = -2\pi^2 e^{at}\sin(\pi y)\sin(\pi x) +\pi^2 e^{at}\sin(\pi y)\sin(\pi x) \Leftrightarrow
a=-\pi^2
\end{equation}
Entonces
\begin{equation}
\boxed{ u(x,y,t) = e^{-\pi^2 t}\sin(\pi x)\sin(\pi y)}
\end{equation}

Comprobamos que es solucin del sistema en las condiciones en los bordes
\begin{split}
&u(x,y,0) = 0 = \sin(\pi x)\sin(\pi y) \\
&u(x,0,t) = u(x,1,t)=u(0,y,t)=u(1,y,t) = 0
\end{split}

## b)


Sea $v(x,y,t)$ de la forma:
\begin{equation}
v(x,y,t) = u(x,y,t) +  \delta\tilde{u}(x,y,t) =u(x,y,t) + \delta e^{bt}\sin(\pi x)\sin(n\pi y)
\end{equation}
donde $b$ lo determinamos a continuacin para que sea solucin del sistema.

Tenemos:
\begin{equation}
\frac{\partial \tilde{u}}{\partial t}(x,y,t) = b e^{bt}\sin(n\pi x)\sin(n\pi y)\quad\quad
\frac{\partial^2 \tilde{u}}{\partial x^2}(x,y,t) = -\pi^2 e^{bt}\sin(\pi x)\sin(n\pi y) \quad\quad
\frac{\partial^2 \tilde{u}}{\partial y^2}(x,y,t) = -(n\pi)^2 e^{bt}\sin(\pi x)\sin(n\pi y) \quad\quad
\end{equation}

Usando la primera condicin:
\begin{equation}
b e^{bt}\sin(n\pi x)\sin(n\pi y) =  -2\pi^2 e^{bt}\sin(\pi x)\sin(n\pi y) +(n\pi)^2 e^{bt}\sin(\pi x)\sin(n\pi y)
\Leftrightarrow b = -2\pi^2+(n\pi)^2 \Leftrightarrow b=\pi^2(n^2-2)
\end{equation}

Entonces:
\begin{equation}
\boxed{v(x,y,t) =  e^{-\pi^2 t}\sin(\pi x)\sin(\pi y) + \delta e^{t\pi^2(n^2-2)}\sin(\pi x)\sin(n\pi y)}
\end{equation}

Comprobamos que se cumple la tercera condicin, adems en la segunda:
\begin{equation}
f(x,y) = v(x,y,0) - \sin(\pi x)\sin(\pi y) = \delta \sin(\pi x)\sin(n\pi y) \Rightarrow |f(x,y)|\leq\delta
\end{equation}
Entonces dado un $\epsilon>0$, escogemos un $\delta:=\epsilon/2$ y entonces $|f(x,y)|<\epsilon$

Ahora, veamos la diferencia de soluciones:
\begin{equation}
v(x,y,T)-u(x,y,T) = \delta  e^{t\pi^2(n^2-2)}\sin(\pi x)\sin(n\pi y) 
\end{equation}

Podemos escoger $n$ para que se cumpla que esta diferencia sea muy grande. Dado $x=1/2, y=1/(2n)$ tenemos:

\begin{split}
&\left|\sin(\pi x)\sin(n\pi y)\right|=1 \\
&\left|\delta e^{t\pi^2(n^2-2)}\right| \xrightarrow{n\rightarrow \infty} \infty
\end{split}

## e)

```python
f = lambda x: np.cos(x)
f_prima = lambda x: -np.sin(x)
```

```python
def calculate_derivative1(f, n):
    h = 2*np.pi/(n-1)
    x = np.linspace(0, 2*np.pi, n)
    f_x = f(x)
    sol = np.zeros(n)
    for i in range(n):
        sol[i-3] = ((-11/6)*f_x[i-3] + 3*f_x[i-2] + (-3/2)*f_x[i-1] + (1/3)*f_x[i])/h
    return sol
```

```python
def calculate_derivative2(f, n):
    h = 2*np.pi/(n-1)
    x = np.linspace(0, 2*np.pi, n)
    f_x = f(x)
    sol = np.zeros(n)
    for i in range(n):
        sol[i-2] = ((1/12)*f_x[i-4] + (-2/3)*f_x[i-3] + (2/3)*f_x[i-1] + (-1/12)*f_x[i])/h
    return sol
```

```python
def calculate_error(f1, f2):
    return np.sum(np.abs(f1-f2))
```

```python
error1, error2 = [], []
x_axis = np.logspace(1, 4)

for i in x_axis:
    n = int(i)

    d_truth = f_prima(np.linspace(0, 2*np.pi, n))
    
    d_approx1 = calculate_derivative1(f, n)
    d_approx2 = calculate_derivative2(f, n)
    
    error1 += [calculate_error(d_approx1, d_truth)]
    error2 += [calculate_error(d_approx2, d_truth)]
    
plt.plot(x_axis, error1, label='Error $\Delta f(x)$')
plt.plot(x_axis, error2, label='Error $\delta f(x)$')
plt.legend()
plt.yscale('log')
plt.ylabel('Error')
plt.xlabel('Number of steps')
plt.show()
```

```python
def f(t):
    s1 = [coefs[i] * np.exp(1j*np.pi*i*t/180) for i in range(L)]
    s2 = [coefs[i].conj() * np.exp(-1j*np.pi*i*t/180) for i in range(1,L)]
    return (np.sum(s1,axis=0) + np.sum(s2,axis=0)).real
```
